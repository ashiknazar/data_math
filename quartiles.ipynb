### **Interquartile Range (IQR)**

- **Definition**: The Interquartile Range (IQR) is a measure of statistical dispersion, or how spread out the middle 50% of the data is. It is the range between the 1st Quartile (Q1) and the 3rd Quartile (Q3).

#### **What Are Quartiles?**

Quartiles divide a data set into four equal parts. These parts contain roughly 25% of the data each, and the three quartiles are:

- **Q1 (First Quartile)**: The median of the lower half of the data (25th percentile). It represents the value below which 25% of the data fall.
- **Q2 (Second Quartile)**: The median of the entire data set (50th percentile). It divides the data into two halves.
- **Q3 (Third Quartile)**: The median of the upper half of the data (75th percentile). It represents the value below which 75% of the data fall.

The IQR is then computed as:

$ \text{IQR} = Q_3 - Q_1 $

Where:
- **$Q_1$** is the 25th percentile (lower quartile)
- **$Q_3$** is the 75th percentile (upper quartile)

#### **Example**

Consider the following data set of exam scores for 7 students:

**Data**: [2, 4, 5, 7, 8, 10, 12]

**Step 1: Arrange the Data in Ascending Order**

The data is already sorted in ascending order:  
**[2, 4, 5, 7, 8, 10, 12]**

**Step 2: Find the Median (Q2)**

The median is the middle value of the data set. Since we have an odd number of data points (7 values), the median is the value at the middle position, which is the 4th value.

- **Median (Q2)** = 7

**Step 3: Find the First Quartile (Q1)**

The first quartile (Q1) is the median of the lower half of the data (everything to the left of the overall median).

- Lower half of the data: [2, 4, 5]
- **Q1** = Median of [2, 4, 5] = 4

**Step 4: Find the Third Quartile (Q3)**

The third quartile (Q3) is the median of the upper half of the data (everything to the right of the overall median).

- Upper half of the data: [8, 10, 12]
- **Q3** = Median of [8, 10, 12] = 10

**Step 5: Calculate the IQR**

Now, we can calculate the **Interquartile Range (IQR)** by subtracting Q1 from Q3:

$ \text{IQR} = Q_3 - Q_1 = 10 - 4 = 6 $

#### **Summary**

- **Data**: [2, 4, 5, 7, 8, 10, 12]
- **Q1 (25th percentile)** = 4
- **Q2 (Median, 50th percentile)** = 7
- **Q3 (75th percentile)** = 10
- **IQR** = 6

The **IQR** indicates that the middle 50% of the data lies between 4 and 10, with a range of 6. It helps identify how spread out the middle values of the data are, which is useful for detecting outliers and understanding data dispersion.
